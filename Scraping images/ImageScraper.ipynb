Converting an image url to image



```python
# import the necessary packages
import numpy as np
import urllib
import cv2
import matplotlib.pyplot as plt
# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
  resp = urllib.request.urlopen(url)
  image = np.asarray(bytearray(resp.read()), dtype="uint8")
  
  image = cv2.imdecode(image, cv2.IMREAD_COLOR)
  image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
	# return the image
  return image
```

```python
IMG_CAPTION = []
```

Web Scraping to scrap specific Category Images

```python
# saved in specific category folder
def Scraper(url, folder):

    if not(os.path.exists(os.path.join('/content/drive/MyDrive/Datasets/Project-Dataset-final', folder))):
      os.mkdir(os.path.join('/content/drive/MyDrive/Datasets/Project-Dataset-final', folder))
    
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    images = soup.find_all('img')
    files = os.listdir(os.path.join('/content/drive/MyDrive/Datasets/Project-Dataset-final',folder))
    category_folder = os.path.join('/content/drive/MyDrive/Datasets/Project-Dataset-final',folder)
    for image in images:
        caption = image['alt']
        link = image['src']
        img_caption = link[65:]+' , '+caption
        if link[-4:] =='.jpg' and link[65:] not in files:
          with open(os.path.join(category_folder,link[65:]), 'wb') as f:
            im = requests.get(link)
            f.write(im.content)
            print('Writing image: ', link[65:])
            IMG_CAPTION.append(img_caption)
             
```

```python
search_category = ['car accidents','domestic violence','fights','injured people','people dead and bleeding',
'fire accidents','women being raped','terrorist']

for tag in search_category:
    count=-1
    while(count<=100):
        count = count+1
        if len(tag.split())==2:
            newtag = tag.split()
            url = f'https://stock.adobe.com/search/images?k={newtag[0]}+{newtag[1]}&search_page={count}'
            # print(url)
            Scraper(url,tag)
        elif len(tag.split())==3:
            newtag = tag.split()
            url = f'https://stock.adobe.com/search/images?k={newtag[0]}+{newtag[1]}+{newtag[2]}&search_page={count}'
            Scraper(url,tag)
            # print(url)
        else:
            url = f'https://stock.adobe.com/search/images?k={tag}&search_page={count}'
            Scraper(url,tag)

```

Saving the image name and the caption in a .txt file

```python
with open(os.path.join('/content/drive/MyDrive/Datasets/Project-Dataset-final','captions.txt'),'w') as f:
          for s in IMG_CAPTION:
            f.write(str(s) +"\n")
```
